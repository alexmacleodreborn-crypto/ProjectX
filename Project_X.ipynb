## Extract Zip File

### Subtask:
Extract the contents of the `/content/Project_X.zip` file into a new directory.


**Reasoning**:
The subtask requires extracting a zip file. I will use the `zipfile` module to extract the contents of `/content/Project_X.zip` into a new directory named `/content/extracted_files`. After extraction, I will list the contents of the new directory to confirm the extraction.



```python
import zipfile
import os

zip_file_path = '/content/Project_X.zip'
extraction_path = '/content/extracted_files'

# Create the extraction directory if it doesn't exist
os.makedirs(extraction_path, exist_ok=True)

# Open the zip file and extract its contents
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_path)

print(f"Contents of '{zip_file_path}' extracted to '{extraction_path}'.")

# List the contents of the extracted directory to verify
print(f"Files in '{extraction_path}':")
for root, dirs, files in os.walk(extraction_path):
    for name in files:
        print(os.path.join(root, name))
    for name in dirs:
        print(os.path.join(root, name))
```

## Create Streamlit App

### Subtask:
Generate a basic Streamlit application that can interact with the extracted files by creating a Python script for the Streamlit app.


**Reasoning**:
I will create a Python script named `streamlit_app.py` within the `/content/extracted_files/` directory, which will contain the code for a basic Streamlit application as instructed. This script will allow users to navigate the extracted files and view their contents.



```python
streamlit_app_path = '/content/extracted_files/streamlit_app.py'

streamlit_code = '''
import streamlit as st
import os

# Define the root directory for exploration
ROOT_DIR = '/content/extracted_files/Project_X'

st.set_page_config(page_title="Project X File Explorer", layout="wide")
st.title("Project X Extracted File Explorer")

# Function to get all directories and files recursively
def get_directory_tree(path):
    tree = {}
    for root, dirs, files in os.walk(path):
        relative_root = os.path.relpath(root, path)
        if relative_root == '.':
            relative_root = ''

        current_level_dirs = sorted([os.path.join(relative_root, d) for d in dirs])
        current_level_files = sorted([os.path.join(relative_root, f) for f in files])

        if relative_root == '': # For the root directory itself
            tree[''] = {'dirs': current_level_dirs, 'files': current_level_files}
        else:
            tree[relative_root] = {'dirs': current_level_dirs, 'files': current_level_files}
    return tree

def display_file_content(filepath):
    try:
        with open(filepath, 'r') as f:
            content = f.read()
            st.subheader(f"Content of {os.path.basename(filepath)}")
            if filepath.endswith(('.py', '.js', '.json', '.txt', '.md')):
                st.code(content, language='python' if filepath.endswith('.py') else 'javascript' if filepath.endswith('.js') else 'json' if filepath.endswith('.json') else 'markdown' if filepath.endswith('.md') else 'text')
            else:
                st.text(content)
    except Exception as e:
        st.error(f"Could not read file: {e}")

# Sidebar for navigation
st.sidebar.header("Navigation")

# Get the directory tree structure
dir_tree = get_directory_tree(ROOT_DIR)

# Get all possible paths (directories and files) for selection
all_paths = []
for root_key, data in dir_tree.items():
    if root_key:
        all_paths.append(root_key)
    # Add files within these directories to the selectable paths
    for f_path in data['files']:
        all_paths.append(os.path.join(root_key, os.path.basename(f_path)) if root_key else os.path.basename(f_path))

# Sort paths for consistent display
all_paths = sorted(list(set(all_paths)))

selected_path_relative = st.sidebar.selectbox(
    "Select a file or directory:",
    [''] + all_paths, # Add an empty string for the root view
    format_func=lambda x: f"./{x}" if x else "./" # Display nice paths
)

full_selected_path = os.path.join(ROOT_DIR, selected_path_relative)

# Main content area
if not selected_path_relative:
    st.subheader(f"Contents of root directory: {ROOT_DIR}")
    st.write("Please select a file or directory from the sidebar.")
    # Display root level contents
    if '' in dir_tree:
        st.write("\n### Directories:")
        for d in dir_tree['']['dirs']:
            st.markdown(f"- **{d}**")
        st.write("\n### Files:")
        for f in dir_tree['']['files']:
            st.markdown(f"- {f}")
elif os.path.isdir(full_selected_path):
    st.subheader(f"Contents of directory: {selected_path_relative}")
    current_dir_info = dir_tree.get(selected_path_relative, {'dirs': [], 'files': []})

    st.write("\n### Subdirectories:")
    if current_dir_info['dirs']:
        for d in current_dir_info['dirs']:
            st.markdown(f"- **{os.path.basename(d)}**")
    else:
        st.write("No subdirectories.")

    st.write("\n### Files:")
    if current_dir_info['files']:
        for f in current_dir_info['files']:
            st.markdown(f"- {os.path.basename(f)}")
    else:
        st.write("No files.")
elif os.path.isfile(full_selected_path):
    display_file_content(full_selected_path)
else:
    st.error("Selected path does not exist or is not a file/directory.")
'''

# Write the Streamlit app content to the file
with open(streamlit_app_path, 'w') as f:
    f.write(streamlit_code)

print(f"Streamlit application script created at: {streamlit_app_path}")
print("You can run this app using: `streamlit run /content/extracted_files/streamlit_app.py`")
```
